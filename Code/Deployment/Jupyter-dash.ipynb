```python
import dash
from jupyter_dash import JupyterDash
from dash import Dash, html, dcc, callback, Output, Input
import plotly.express as px
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
#Crear la aplicación
app = JupyterDash(__name__)

app.layout = html.Div(
    [
        html.H1('Mi primera APP'),
        html.Div('Esta es mi primera  aplicación .... espero no se vea tan fea'),
        dcc.Graph(id='Mi primer gráfico'),
        html.Label(["colorscale",dcc.Dropdown(id='colorscale-dropdown',
                                               clearable=False,
                                               value='plasma',
                                               options=[{'label': c, 'value': c}
                                                        for c in px.colors.named_colorscales()])
                                                        ])]
)

@app.callback(Output('Mi primer gráfico', 'figure'),[Input("colorscale-dropdown", "value")])
def update_figure(colorscale):
    return px.scatter(df, x="total_bill", y="tip", color="size",color_continuous_scale=colorscale,render_mode="webgl", title="Tips")


app.run_server(mode='inline')
```

```python
from dash import Dash, html, dcc, callback, Output, Input
import plotly.express as px
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gapminder_unfiltered.csv')

app = Dash(__name__)

app.layout = html.Div([
    html.H1(children='Title of Dash App', style={'textAlign':'center'}),
    dcc.Dropdown(df.country.unique(), 'Canada', id='dropdown-selection'),
    dcc.Graph(id='graph-content')
])

@callback(
    Output('graph-content', 'figure'),
    Input('dropdown-selection', 'value')
)
def update_graph(value):
    dff = df[df.country==value]
    return px.line(dff, x='year', y='pop')

if __name__ == '__main__':
    app.run_server(debug=True)
```
